# 지역화폐가맹점 데이터

```python
import pandas as pd
df1 = pd.read_csv('지역화폐 가맹점_20240331.csv',encoding ='cp949')
df1
```

```python
# 1. 시군명, 업종명(종목명), 업종코드, 휴폐업상태, 휴페업상태 코드, 폐업일자 남기고 다른 column 삭제
df1 = df1.drop(['상호명', '소재지도로명주소','소재지지번주소','우편번호', '위도', '경도', '사업자등록번호', '가맹점번호'], axis = 1)
df1
```

# 데이터 전처리

```python
# 데이터의 구조와 타입(변수 속성)
print('데이터1의 구조와 타입 : ')
print(df1.info())
```

```python
# 데이터의 결측치 확인
print(('데이터1의 결측치 : '))
df1.isnull().sum()
```

```python
# 휴폐업 상태 코드는 뭔지 잘 모름. 찾아보니 계속 사업자 1, 휴업자 2, 폐업자 3 그렇다면 휴폐업 상태를 버려도? 코드만 보고 분석 가능
#  Nan을 1로 치환 (문의해봄)
df1['휴폐업상태코드'] = df1['휴폐업상태코드'].fillna(1).astype(int)
df1['휴폐업상태'] = df1['휴폐업상태'].fillna('계속사업자')

df1
```

```python
#업종명 검토

# 시군명과 업종명으로 crosstab 테이블 생성
crosstab_table = pd.crosstab(index=df1['업종명(종목명)'], columns=df1['시군명'])

# 모든 행과 열을 표시하도록 설정
pd.set_option('display.max_rows', None)   # 모든 행 표시
pd.set_option('display.max_columns', None) # 모든 열 표시
pd.set_option('display.width', None)       # 자동 너비 조정
pd.set_option('display.max_colwidth', None) # 모든 열 너비 표시

# crosstab 테이블 출력
crosstab_table
```

```python
# 백화점, 대형마트, 기업형 슈퍼마켓(SSM), 프렌차이즈 직영점 및 유흥 사행업소는 사용이 제한 : 삭제
df1_cleaned = df1[~df1['업종명'].isin(['백화점', '대형마트', '기업형 슈퍼마켓(SSM), '])]

# 결과를 확인합니다.
print(df1_cleaned)
# 비슷한 부류 묶기
```

```python
# 업종명 - 업종코드(업종 종류를 분류 할 수 있음)
df1['업종명(종목명)'].unique()
```

```python
# 업종명 별 업종코드
df1.pivot_table(index = '업종명(종목명)', values = '업종코드').reset_index()
```

```python
#3-1. 시 별로 상점 총 개수 구하기
업종별_개수 = df1.groupby(['시군명', '업종명(종목명)']).size().reset_index(name='개수')

# 결과 출력
print("시군명별 업종명(종목명) 개수:")
업종별_개수
```

```python
# 3-1. 시각화
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import plotly.express as px
mpl.rc('font', family = 'AppleGothic')

# 시군명별 총 업종 개수 구하기
시군별_업종수 = 업종별_개수.groupby('시군명')['개수'].sum().reset_index()

# 가장 많은 업종을 가진 상위 5개 시군명 추출
상위5개_시군명 = 시군별_업종수.nlargest(5, '개수')['시군명']

# 상위 5개 시군명에 대해 업종별 개수 필터링
상위5개_업종별_개수 = 업종별_개수[업종별_개수['시군명'].isin(상위5개_시군명)]

시군별_업종수 #(상점 수)
```

```python
시군별_업종수_최종 = 시군별_업종수.sort_values(by = '개수', ascending = False).head(5)
시군별_업종수_최종
```

```python
# 시군별_업종수_최종 시각화
plt.figure(figsize=(12, 8))
sns.barplot(data=시군별_업종수, x='시군명', y='개수')
plt.title('시군별 상점수')
plt.xlabel('시군명')
plt.ylabel('싱점수')
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#5-1. 계속사업자의 업종 비율

# '휴폐업상태코드'가 1인 계속사업자 데이터 필터링
계속사업자_df = df1[df1['휴폐업상태코드'] == 1]

# 업종명별 개수 세기
업종별_개수 = 계속사업자_df['업종명(종목명)'].value_counts()

# 상위 10개 추출
상위_10개 = 업종별_개수.head(10)

상위_10개
```

```python
# 비율 계산
총_계속사업자_업종수 = 업종별_개수.sum()
비율 = 상위_10개 / 총_계속사업자_업종수 * 100

# 시각화
plt.figure(figsize=(8, 8))

# Seaborn 색상 팔레트 사용
colors = sns.color_palette('pastel')[0:10] 

plt.pie(비율, labels=비율.index, autopct='%1.1f%%', colors=colors)
plt.title('계속사업자 업종명별 상위 10개 비율')
plt.axis('equal')# 원이 정원으로 보이도록
plt.tight_layout()
plt.show()
```

```python
# 7. 폐업한 업종을 분류해서 어떤 업종은 지역화폐를 사용해도 폐업할 가능성이 높다는 결론을 도출해낼 수도 있음

# 폐업한 업종의 상위 5개
# 폐업자 코드 3
closed = df1[df1['휴폐업상태코드']== 3]
closed_store = closed['업종명(종목명)'].value_counts()
closed_store = pd.DataFrame(closed_store.head(10)).reset_index()

closed_store = closed_store.rename(columns = {'index':'업종명(종목명)', '업종명(종목명)':'개수'})
closed_store
```

```python
# 7. 폐업한 업종 시각화
plt.figure(figsize=(8, 6))

# Seaborn 색상 팔레트 사용
colors = sns.color_palette('pastel')[0:10] 

plt.pie(closed_store['개수'], labels=closed_store['업종명(종목명)'], autopct='%1.1f%%', startangle=240, colors = colors)
plt.title('폐업한 업종 비율')
plt.axis('equal')  # 원을 동그랗게 유지
plt.tight_layout()
plt.show()
```

```python
# 11. 22년데이터에서 새로 추가된 4천개의 데이터만 뽑아 최근 2년간 어떤 업종이 유행했는지도 분석 가능
df_2022 = pd.read_csv('지역화폐가맹점현황_20221231.csv',encoding ='cp949')
df_2024 = pd.read_csv('지역화폐 가맹점_20240331.csv',encoding ='cp949')

# 22년도와 24년도 데이터에서 겹치는 부분을 '겹치는 항목'으로 지정
merged_df = df_2024.merge(df_2022, on=['시군명', '업종명(종목명)', '업종코드', '휴폐업상태', '휴폐업상태코드', '폐업일자'], how='left', indicator=True)

# 겹치는 항목만 필터링
겹치는_항목 = merged_df[merged_df['_merge'] == 'both'].drop(columns=['_merge'])

# 24년도 데이터에서 겹치는 항목을 제외
df_2024_제외 = merged_df[merged_df['_merge'] == 'left_only'].drop(columns=['_merge'])

# 결과 출력
#print("겹치는 항목:")
#print(겹치는_항목)

print("\n24년도 데이터에서 겹치는 항목을 제외한 결과:")
df_2024_제외
```
