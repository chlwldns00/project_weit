```python
import pandas as pd
df_visit = pd.read_csv('지역별 방문자 수.csv', encoding='cp949')
df_visit
```

```python
#컬럼명 변경
df_visit = df_visit.rename(columns={'기초지자체명':'시군명', '기초지자체 방문자 수':'관광객수', '기초지자체 방문자 비율':'관광객비율'})
```

```python
#전처리
# 시군명 통합
df_visit['시군명'] = df_visit['시군명'].replace({'고양시 덕양구': '고양시','고양시 일산동구': '고양시','고양시 일산서구': '고양시'})
df_visit['시군명'] = df_visit['시군명'].replace({'부천시 소사구': '부천시','부천시 오정구': '부천시','부천시 원미구': '부천시'})
df_visit['시군명'] = df_visit['시군명'].replace({'성남시 분당구': '성남시','성남시 수정구': '성남시','성남시 중원구': '성남시'})
df_visit['시군명'] = df_visit['시군명'].replace({'수원시 권선구': '수원시','수원시 영통구': '수원시','수원시 장안구': '수원시', '수원시 팔달구':'수원시'})
df_visit['시군명'] = df_visit['시군명'].replace({'안산시 단원구': '안산시','안산시 상록구': '안산시'})
df_visit['시군명'] = df_visit['시군명'].replace({'안양시 동안구': '안양시','안양시 만안구': '안양시'})
df_visit['시군명'] = df_visit['시군명'].replace({'용인시 기흥구': '용인시','용인시 수지구': '용인시','용인시 처인구': '용인시'})
```

```python
# 관광객수 집계
df_visit = df_visit.groupby('시군명', as_index=False).agg({'관광객수': 'sum'})
df_visit['관광객수'] = df_visit['관광객수'].astype(int)
df_visit
```

```python
df_visit.to_csv('관광객수.csv')
df_visit
```

```python
df_people = pd.read_excel('경기도 인구통계.xlsx')
df_people = df_people[['시군명','총인구']]
df_people
```

```python
인구수 = pd.merge(df_visit,df_people, on ='시군명', how = 'left')
인구수.head()
```

```python
data3 = pd.read_csv('Data3_전처리.csv')

정민군집화데이터 = pd.merge(data3,인구수, on ='시군명', how = 'left')
정민군집화데이터.to_csv('정민군집화데이터.csv')
```
